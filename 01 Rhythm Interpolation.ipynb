```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v0 = np.zeros(12)
v0[1] = 0.5
v0[6] = 0.5

v1 = np.zeros(12)
v1[3] = 0.5
v1[9] = 0.5
```

```python
plt.plot(v0)
plt.plot(v1+0.01);
```

```python
T = np.zeros((12, 12))
T[1, 3] = 0.5
T[6, 9] = 0.5
```

```python
plt.matshow(T);
```

```python
def interp(v0, v1, T, p):
    vp = np.zeros_like(v0)
    for i in range(len(vp)):
        js = np.where(T[i] > 0)[0]
        for j in js:
            j_p = int(p * j + (1-p)*i)
            vp[j_p] += T[i, j]
            
    return vp
```

```python
for i, p in enumerate(np.linspace(0, 1, 5)):
    plt.plot(interp(v0, v1, T, p)+0.01*i, label=p)
    
plt.legend()
plt.show()
```
